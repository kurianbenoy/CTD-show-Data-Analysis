```python
import re
def extracting_links(text):
    text = re.sub(r'\n',' ',text)
    l = re.findall(r'http[s]*://[a-zA-Z0-9./-]+',text)
    return l
```

```python
eps_desc['links'] = eps_desc['description'].apply(lambda x : extracting_links(x))
```
